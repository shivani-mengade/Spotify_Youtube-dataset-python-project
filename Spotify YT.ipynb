```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#IMPORT FILE
data = pd.read_csv(r'C:\Users\HP\Downloads\Spotify_Youtube Dataset.csv', encoding='utf-8')
```

```python
data.shape
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(columns = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace = True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
#fill values
data['Likes'] = data['Likes'].fillna(0)
data['Comments'] = data['Comments'].fillna(0)
```

```python
data.dropna(inplace = True)
data.isnull().sum()
```

```python
#TOP 10 ARTIST WITH HIGHEST VIEWS ON YOUTUBE
Artist_grp = data.groupby('Artist')['Views'].sum()
Artist_s = Artist_grp.sort_values(ascending = False)
Artist_s.head(10)
```

```python
#TOP 10 TRACKS WITH HIGHEST STREAM ON SPOTIFY
s = data[['Track', 'Stream']]
most_viewed = s.sort_values(by= ['Stream'], ascending = False).head(10)
most_viewed
```

```python
#TRACK WITH LOWEST STREAM
s = data[['Track', 'Stream']]
least_streamed = s.sort_values(by= ['Stream'], ascending = True).head(5)
least_streamed
```

```python
#COMMON ALBUM TYPE ON SPOTIFY?ALBUM TO EACH  ALBUM TYPE?
data.Album_type.unique()
a = data['Album_type'].value_counts()
a
```

```python
plt.pie(a, labels = a.index, autopct ="''%1.1f%%")
plt.show()
```

```python
# VIEWS, LIKES, COMMENTS COMPARED of DIFF ALBUM TYPES?
df = data.groupby('Album_type')[['Views', 'Likes', 'Comments']].mean()
df  = df.reset_index()
df
df.m = pd.melt(df, id_vars = 'Album_type', var_name = 'Attribute', value_name = 'Total')
df.m
```

```python
sns.barplot(x = 'Album_type', y= 'Total', hue = 'Attribute', data= df.m );
plt.show()
```

```python
#TOP 5 YT CHANNELS BASED ON VIEWS
c_v = data.groupby('Channel')['Views'].sum().sort_values(ascending = False).head().reset_index()
c_v
```

```python
#sns.set_style('whitegrid')
sns.barplot(x = 'Views', y ='Channel', data = c_v)
plt.title('Top 5 Channels')
plt.show()
```
